# Apply .apply

```python
df['new_col_square_d'] = df['d'].apply(lambda x: x**2)
```

```python
df
```

```python
df['new_col_square_d_log'] = df['d'].apply(lambda x: np.log(x))
```

```python
df
```

```python
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
test(2)
```

```python
df['new_col_square_d_log_1'] = df['d'].apply(test)
```
