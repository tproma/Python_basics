```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

# Seaborn

```python
df = sns.load_dataset('tips')
```

```python
df
```

# Seaborn Relational Plotting 

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'smoker')
```

```python
df
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'day')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'day', style = 'smoker')
```

# hue -> based on color , style -> based on shape , differentiates the group

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'day', style = 'smoker', size='size' )
```
